```python
year_list = ["1973", "1974","1975","1976","1977","1978",
             "1979","1980","1981","1982","1983","1984",
             "1985","1986","1987","1988","1989","1990",
             "1991","1992","1993","1994","1995","1996",
             "1997","1998","1999","2000","2001","2002",
             "2003","2004","2005","2006","2007","2008",
             "2009","2010","2011","2012","2013","2014",
             "2015","2016","2017","2018","2019","2020",
             "2021"]
year_mean_mean = []
for i in year_list:
    year_mean_mean.append(df_mean_temp[df_mean_temp['날짜'].str.startswith(i)])
year_max_mean = []
for i in year_list:
    year_max_mean.append(df_max_temp[df_max_temp['날짜'].str.startswith(i)])
year_min_mean = []
for i in year_list:
    year_min_mean.append(df_min_temp[df_min_temp['날짜'].str.startswith(i)])
```

```python
year_mean_mean[48]
year_max_mean[48]
year_min_mean[48]
```

```python
year_mean_mean_list = []
year_min_mean_list = []
year_max_mean_list = []

for i in range(0, 49):
    year_mean_mean_list.append(year_mean_mean[i].mean())
    
for i in range(0, 49):
    year_min_mean_list.append(year_max_mean[i].mean())
    
for i in range(0, 49):
    year_max_mean_list.append(year_min_mean[i].mean())
```

```python
df_meantemp = pd.DataFrame(year_mean_mean_list)
df_maxtemp = pd.DataFrame(year_max_mean_list)
df_mintemp = pd.DataFrame(year_min_mean_list)
```

```python
df_result = pd.concat([df_meantemp, df_maxtemp], axis=1)
df_result
```

```python
df_result = pd.concat([df_result, df_mintemp], axis = 1)
df_result
```

```python
sns.lineplot(data = df_result)
plt.show()
```

```python
df_result = df_result.rename(columns={0 : "Year"})
df_result
```
